## E2.1
A single input neuron has a weight of 1.3 and a bias of 3.0. What possible
kinds of transfer functions, from Table 2.1, could this neuron have, if its
output is given below. In each case, give the value of the input that would
produce these outputs.
i. 1.6
ii. 1.0
iii. 0.9963
iv. -1.0

```python
import numpy as np
w =1.3
b=0.3
    
def get_netinput(w,b,input):
    return (w*input+b)

def linear_transfer_function(netinput):
    return netinput

def sigmoid_transfer_function(netinput):
    return (1/(1+np.exp(-netinput)))

def hardlimit_transfer_function(netinput):
    if netinput>=0:
        return 1.0
    else:
        return 0

def symmeterical_hardlimit_transfer_function(netinput):
    if netinput>=0:
        return 1.0
    else:
        return -1.0

    
def exercise_i():
    input=1.0
    print("weight: ", w)
    print("bias: ", b)    
    print("When input is ", input, " and linear transfer function is applied, then the output is :")
    print(linear_transfer_function(get_netinput(w,b,input)))

def exercise_ii():
    input=0.4
    print("weight: ", w)
    print("bias: ", b)    
    print("When input is ", input, " and hard-limit transfer function is applied, then the output is :")
    print(hardlimit_transfer_function(get_netinput(w,b,input)))

def exercise_iii():
    input=4.07
    print("weight: ", w)
    print("bias: ", b)    
    print("When input is ", input, " and sigmoid transfer function is applied, then the output is :")
    print(round(sigmoid_transfer_function(get_netinput(w,b,input)),4))

def exercise_iv():
    input=-0.4
    print("weight: ", w)
    print("bias: ", b)    
    print("When input is ", input, " and sigmoid transfer function is applied, then the output is :")
    print(symmeterical_hardlimit_transfer_function(get_netinput(w,b,input)))
    
exercise_i()
print("\n")
exercise_ii()
print("\n")
exercise_iii()
print("\n")
exercise_iv()

```
